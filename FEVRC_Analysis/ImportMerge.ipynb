```python
''' Drop Unneeded Columns and Rename Needed Columns if Needed'''
fueleconomy = fueleconomy.iloc[:, :6]
registrations = registrations.iloc[:, :4]
consumption = consumption.iloc[:, :2]
consumption.columns = ["Date", "Gasoline_Consumption(Millions of Gallons)"]
```

```python
consumption.info()
print("\n")
fueleconomy.info()
print("\n")
registrations.info()
```

# Transform Registrations

```python
# create new column of total registrations
registrations['TotalRegistrations'] = registrations['Motorcycle_Registrations'] + registrations['Automobile_Registrations']
registrations['TotalRegistrations'] = registrations['TotalRegistrations'].round(0)
registrations['Year'] = registrations['Year'].dt.year
registrations['Year'] = registrations['Year'].astype(int)
```

# Verify Data

```python
consumption.info()
print("\n")
fueleconomy.info()
print("\n")
registrations.info()
```

# Merge

```python
# Merge all three datasets together
merged_df = pd.merge(fueleconomy, registrations, on='Year', how='inner')
merged_df = pd.merge(merged_df, consumption, on='Year', how='inner')
merged_df
```
